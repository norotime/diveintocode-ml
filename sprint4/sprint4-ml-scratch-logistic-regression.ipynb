```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ロジスティック関数
def logistic_function(x):
    return 1/ (1+ np.exp(-x))
```

```python
#linspace 指定された間隔で等間隔の数値を返す。
#区間[start、stop]にわたって計算された、num個の等間隔サンプルを返す。間隔の終点はオプションで除外できる。
x = np.linspace(-10,10,1000)
plt.plot(x, logistic_function(x))
plt.title("LogisticFunction")
plt.xlabel("x")
plt.ylabel("y")
```
