```python
thickness = 0.08
print("thickness's id", id(thickness))
folded_thickness = thickness
print("folded_thickness's id", hex(id(folded_thickness)))
print("foled_thickness :", folded_thickness)

for _ in range(10):
    folded_thickness = folded_thickness*2
    print("folded_thickness's id", hex(id(folded_thickness)))
    print("foled_thickness :", folded_thickness)
```
